```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import mean_squared_log_error
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df.drop(['ID','Model','Category'],axis = 1,inplace = True)
```

```python
df.replace(to_replace='-',value=np.nan,inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer().fit(df[['Levy']])
data_array = imputer.transform(df[['Levy']])
```

```python
data_array
```

```python
df.drop(['Levy'],axis = 1, inplace = True)
```

```python
df['Levy'] = data_array
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder().fit(df[['Manufacturer','Leather interior','Fuel type','Gear box type','Drive wheels','Wheel','Color']])
onehot_data = enc.transform(df[['Manufacturer','Leather interior','Fuel type','Gear box type','Drive wheels','Wheel','Color']])
```

```python
onehot_data = pd.DataFrame(onehot_data.todense())
```

```python
data = pd.concat([df,onehot_data],axis = 1)
```

```python
data
```

```python
data.drop(['Manufacturer',
           'Leather interior','Fuel type','Gear box type','Drive wheels','Wheel','Color','Cylinders'],axis = 1, inplace = True)
```

```python
data['Doors'].unique()
```

```python
data['Doors'].replace(to_replace='>5',value=5,inplace=True)
```

```python
data
```

```python
X = data.drop(['Price'],axis =1)
y = data[['Price']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)
scal_data = scaler.transform(X)
```

```python
X_columns = data.columns
len(X_columns)
```

```python
dem = pd.DataFrame(data=scal_data,columns=X_columns[1:])
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=60)
final = pca.fit(dem)
final_data = final.transform(dem)
```

```python
#final_data.shape
```

```python
TEST_SIZE = 0.2
RAND_STATE = 10
```

```python
from sklearn.model_selection import train_test_split #split
from sklearn.metrics import accuracy_score #metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_data, y, test_size = TEST_SIZE, random_state=RAND_STATE)
```

```python
from sklearn.ensemble import RandomForestRegressor
#from sklearn.datasets import make_regression
#X, y = make_regression(n_features=4, n_informative=2,random_state=10, shuffle=False)
regr = RandomForestRegressor(max_depth=15, random_state=0)
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
```

```python
y_pred
```

```python
#y_test = np.array(y_test)
```

```python
y_test = abs(y_test)
```

```python
regr.score(X_train,y_train)
```

```python
y_pred = regr.predict(X_test)
y_pred = abs(y_pred)
score = np.sqrt(mean_squared_log_error(y_test,y_pred))
score
```
